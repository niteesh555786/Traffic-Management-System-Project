# **Forecasting**

## **1: Start with a Naive Approach**

Next expected point is equal to the last observed point is called Naive Method

```
dd= np.asarray(train['AIT-Dighi'])
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
plt.figure(figsize=(12,8))
plt.plot(train.index, train['AIT-Dighi'], label='Train')
plt.plot(test.index,test['AIT-Dighi'], label='Test')
plt.plot(y_hat.index,y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best')
plt.title("Naive Forecast")
plt.show()
```

## 2:  Simple Average

```
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['AIT-Dighi'].mean()
plt.figure(figsize=(12,8))
plt.plot(train['AIT-Dighi'], label='Train')
plt.plot(test['AIT-Dighi'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
plt.show()
```

```
from sklearn.metrics import mean_squared_error
rms = np.sqrt(mean_squared_error(test['AIT-Dighi'], y_hat_avg.avg_forecast))
print(rms)
```

## 3  Moving Average

we will take the average of the prices for last few time periods only. Obviously the thinking here is that only the recent values matter. Such forecasting technique which uses window of time period for calculating the average is called Moving Average technique. Calculation of the moving average involves what is sometimes called a “sliding window” of size n.

```
y_hat_avg = test.copy()
y_hat_avg['moving_avg_forecast'] = train['AIT-Dighi'].rolling(60).mean().iloc[-1]
plt.figure(figsize=(16,8))
plt.plot(train['AIT-Dighi'], label='Train')
plt.plot(test['AIT-Dighi'], label='Test')
plt.plot(y_hat_avg['moving_avg_forecast'], label='Moving Average Forecast')
plt.legend(loc='best')
plt.show()
```

## 4 – Holt’s Linear Trend method

Each Time series dataset can be decomposed into it’s componenets which are Trend, Seasonality and Residual. Any dataset that follows a trend can use Holt’s linear trend method for forecasting.

```
import statsmodels.api as sm
sm.tsa.seasonal_decompose(train['AIT-Dighi']).plot()
result = sm.tsa.stattools.adfuller(train['AIT-Dighi'])
plt.show()
```

## 5-ARIMA

It stand for Autoregressive Integrated Moving average. While exponential smoothing models were based on a description of trend and seasonality in the data, ARIMA models aim to describe the correlations in the data with each other. An improvement over ARIMA is Seasonal ARIMA. It takes into account the seasonality of dataset just like Holt’ Winter method. 

```
y_hat_avg = test.copy()
fit1 = sm.tsa.statespace.SARIMAX(train['AIT-Dighi']).fit()
y_hat_avg['SARIMA'] = fit1.predict(start="2018-1-1 01:00:00", end="2018-03-22 00:00:00", dynamic=True)
plt.figure(figsize=(16,8))
plt.plot( train['AIT-Dighi'], label='Train')
plt.plot(test['AIT-Dighi'], label='Test')
plt.plot(y_hat_avg['SARIMA'], label='SARIMA')
plt.legend(loc='best')
plt.show()
```
